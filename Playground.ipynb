```python
import numpy as np
import matplotlib.pyplot as plt
# sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate a range of values for x
x = np.linspace(-10, 10, 100)
x2 = np.linspace(-12, 10, 100)

# Calculate the sigmoid function for each x
y = sigmoid(x)
y2 = sigmoid(x2)

# Plot the sigmoid function
plt.plot(x, y)
plt.plot(x2, y2)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('S(x)')
plt.grid(True)
plt.show()
```
